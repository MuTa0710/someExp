## AI给出的简单实现

```python
import numpy as np  # 导入 NumPy 库，用于进行高效的数值计算和数组操作

def simple_kmeans(data, k, max_iterations=100):  # 定义一个函数，接收数据集、簇的数量(k)以及最大迭代次数
    
    num_samples = data.shape[0]  # 获取数据集中的样本总数（即矩阵的行数）
    
    num_features = data.shape[1]  # 获取每个样本的特征数量（即矩阵的列数）
    
    random_indices = np.random.choice(num_samples, k, replace=False)  # 从所有样本索引中，随机且不重复地抽出 k 个索引
    
    centroids = data[random_indices]  # 根据随机抽出的索引，从数据集中取出对应的样本，作为初始的 k 个聚类中心
    
    cluster_assignments = np.zeros(num_samples)  # 初始化一个数组，用于记录每个样本所属的簇的编号，初始都设为 0
    
    for iteration in range(max_iterations):  # 开始迭代，最多执行 max_iterations 次
        
        # --- 第一步：将每个点分配给最近的聚类中心 ---
        
        for i in range(num_samples):  # 遍历数据集中的每一个样本
            
            current_point = data[i]  # 获取当前正在处理的第 i 个样本
            
            # 计算当前样本到所有 k 个聚类中心的欧氏距离的平方（不开根号不影响比较大小，且计算更快）
            squared_distances = np.sum((centroids - current_point) ** 2, axis=1)  
            
            closest_centroid_index = np.argmin(squared_distances)  # 找到距离最小的那个聚类中心的索引
            
            cluster_assignments[i] = closest_centroid_index  # 将当前样本分配给距离它最近的那个聚类中心
            
        # --- 第二步：重新计算聚类中心 ---
        
        new_centroids = np.zeros((k, num_features))  # 创建一个全 0 矩阵，用于存放这一轮计算出的新的聚类中心
        
        for j in range(k):  # 遍历所有的 k 个簇
            
            # 找出所有被分配到第 j 个簇的样本点
            points_in_cluster_j = data[cluster_assignments == j]  
            
            if len(points_in_cluster_j) > 0:  # 如果这个簇里面有样本点（为了防止除以零的错误）
                
                # 计算这个簇里所有样本点在各个特征维度上的平均值，作为新的聚类中心
                new_centroids[j] = np.mean(points_in_cluster_j, axis=0)  
                
            else:  # 如果这个簇里没有任何样本点（极端情况，比如被其他中心抢光了）
                
                new_centroids[j] = centroids[j]  # 保持原聚类中心不变，或者也可以选择重新随机初始化
                
        # --- 第三步：检查是否满足结束条件（收敛） ---
        
        if np.all(centroids == new_centroids):  # 如果这一轮计算出的新聚类中心和老聚类中心完全一模一样
            
            break  # 说明聚类结果已经稳定，不再变化，提前跳出迭代循环
            
        centroids = new_centroids  # 如果中心还在变化，就把新计算出的中心赋值给 centroids，准备进行下一次迭代

    return cluster_assignments, centroids  # 返回最终每个样本所属的簇编号，以及最终的聚类中心坐标

# ==========================================
# 下面是一个简单的测试用例，用来验证上述代码
# ==========================================

if __name__ == "__main__":  # 如果直接运行这个脚本文件
    
    # 手动创建一些二维数据点，明显可以分为两群
    test_data = np.array([[1.0, 2.0], [1.5, 1.8], [5.0, 8.0], [8.0, 8.0], [1.0, 0.6], [9.0, 11.0]])  
    
    k_value = 2  # 我们假设要把这些数据分成 2 个簇
    
    # 调用我们手写的 kmeans 函数
    final_labels, final_centroids = simple_kmeans(test_data, k_value)  
    
    print("每个样本的聚类标签 (所属簇的编号):")  # 打印提示信息
    print(final_labels)  # 打印每个样本分配的标签，例如 [0, 0, 1, 1, 0, 1]
    
    print("\n最终的聚类中心坐标:")  # 打印提示信息
    print(final_centroids)  # 打印最终确定下来的 k 个聚类中心的位置
```

```python
import numpy as np

def Kmeans(data, k, iteration=100):

    sample_num, feature_dim = data.shape

    center_id = np.random.choice(sample_num, k, replace=False)
    center_point = data[center_id]

    cata = np.zeros(sample_num)

    for _ in range(iteration):
        for i in range(sample_num):
            distance = np.sum((center_point - data[i]) ** 2, axis=1)
            mindistance_id = np.argmin(distance)
            cata[i] = mindistance_id
        new_center_point = np.zeros((k, feature_dim))
        for j in range(k):
            temp_sample = data[j == cata]
            if len(temp_sample) > 0:
                mean_cent = np.mean(temp_sample, axis=0)
                new_center_point[j] = mean_cent
            else:
                new_center_point[j] = center_point[j]
        if np.all(center_point == new_center_point):
            break
        center_point = new_center_point

    return cata, center_point

            
```
