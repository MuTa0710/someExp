# RoPE

## Code

```python
import torch
import torch.nn as nn

def Gen_freq(dim, theta):
    freq = 1.0 / (theta ** (torch.arange(0, dim, 2)[:(dim // 2)] / dim))
    return freq

class RoPE(nn.Module):
    def __init__(self, dim, theta=10000) -> None:
        super().__init__()
        self.dim = dim
        self.theta = theta
        self.Gen_freq = Gen_freq(self.dim, self.theta)

print(Gen_freq(8, 10000))
```
