# RoPE

## Code

```python
import torch
import torch.nn as nn

def Gen_freq(dim, theta, seqlen):
    freq = 1.0 / (theta ** (torch.arange(0, dim, 2)[:dim // 2]).float() / dim)

    # theta_i shape: dim//2

    # freq = torch.tensor([1.0 / theta ** (2 * i / dim) for i in range(dim // 2)])# shape dim // 2
    # theta^(-2 * i / dim)
    index = torch.arange(seqlen).float()
    freqs = torch.outer(index, freq).float()# seqlen dim//2

    freq_cis = torch.polar(torch.ones_like(freqs), freqs)
    #[cos(m * theta_i) + j * sin(m * theta_i)]
    return freq_cis # seqlen dim//2

class RoPE(nn.Module):
    def __init__(self, dim, seqlen, theta=10000) -> None:
        super().__init__()
        self.dim = dim
        self.theta = theta
        self.seqlen = seqlen
        self.Gen_freq = Gen_freq(self.dim, self.theta, self.seqlen)
        self.cos = torch.cos(self.Gen_freq).repeat_interleave(2, dim = -1)
        self.sin = torch.sin(self.Gen_freq).repeat_interleave(2, dim = -1)


print(Gen_freq(8, 10000, 2))
```
