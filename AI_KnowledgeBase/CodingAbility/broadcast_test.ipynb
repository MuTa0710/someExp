```python
mat = torch.ones(8, 8)
mat
```

```python
mat = torch.triu(mat)
mat
```

```python
mat = torch.triu(mat, diagonal=1)
mat
```

```python
mask = torch.full((8, 8), float("-inf"))
mask
```

```python
mask = torch.triu(mask, diagonal=1)
mask
```

```python
masked = mask + mat
masked
```

```python
scores = torch.softmax(masked, dim=-1)
scores
```
